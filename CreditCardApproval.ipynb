```python
# Importing the libraries..

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading the datsets

df  = pd.read_csv("/content/Credit_card.csv")
df1 = pd.read_csv("/content/Credit_card_label.csv")
```

```python
# Merging the data set
df_credit = pd.merge(df,df1,on = "Ind_ID")
```

# **Encoding**

```python
# converting non numeric to numerical variables.

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df_credit:
  if df_credit[col].dtype=='object':
    df_credit[col]=le.fit_transform(df_credit[col])
```

```python
df_credit.info()
```

```python
df_credit.head()
```

```python
df_credit.shape
```

```python
df_credit.isnull().sum()
```

```python
# Initial hypothesis:There is no significant association between each categorical feature and the target variable ('Label').
# so here i am using Chi-Square test.

x = df_credit.drop(['label'],axis=1)    # independent varaibles
y = df_credit[['label']]       # target variable


```

```python
# split the data into train and test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state= 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# chi2 return 2 values -- Fscore and p-value

from sklearn.feature_selection import chi2
f_p_values = chi2(x_train,y_train)
f_p_values
```

```python
p_values = pd.Series(f_p_values[1])
p_values.index = x_train.columns
p_values
```

```python
# sorted the p-values
p_values.sort_index(ascending= True)
```

Here I have observed that  'Annual_income', 'Employed_Exp', and 'Type_Income' are potentially significant features in relation to the target variable based on the chi-squared test.








# **Feature Scaling...**

```python
# Importing MinMaxScaler

from sklearn.preprocessing import MinMaxScaler
scaler  = MinMaxScaler(feature_range= (0,1))
scale_x_train = scaler.fit_transform(x_train)
scale_x_test = scaler.fit_transform(x_test)
print(scale_x_train.shape)
print(scale_x_test.shape)
```

Applying ML algorithm to the pre-processed columns and finding which of these algorithms is best suitable

# **Date Modeling**

**1) Logistic Regression.**

```python
from sklearn.linear_model import LogisticRegression
l_reg = LogisticRegression()
l_reg.fit(scale_x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred = l_reg.predict(scale_x_test)

print("Logistic Regression Training set:",l_reg.score(x_train,y_train))
print("Logistic Regression Testing set:",l_reg.score(x_test,y_test))
print("Accuracy of Logistic regression: ",accuracy_score(y_test,y_pred))
print("Confusion Matrix of Logistic Regression:","\n",confusion_matrix(y_test,y_pred))


```

**Cross Validation**

```python
from sklearn.model_selection import cross_val_score,ShuffleSplit

log_model =  LogisticRegression()
ssplit = ShuffleSplit(n_splits = 10,test_size = 0.20,random_state = 0)
cross_score = cross_val_score(log_model,x,y, cv = ssplit)
print(cross_score)
print(np.mean(cross_score))
```

**2) Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)

y_pred_rf = rf.predict(x_test)

print("Random Forest Training set: ",rf.score(x_train,y_train))
print("Random Forest Testing set: ",rf.score(x_test,y_test))
print("Accuracy of Random Forest Classifier: ",accuracy_score(y_test,y_pred_rf))

print("Confusion Matrix:","\n",confusion_matrix(y_test,y_pred_rf))
```

**Cross Validation (RF)**

```python
from sklearn.model_selection import ShuffleSplit,cross_val_score
rf_model = RandomForestClassifier()
ssplit = ShuffleSplit(n_splits=10,test_size = 0.20)
cross_score = cross_val_score(rf_model,x,y,cv=ssplit)

print(cross_score)
print(np.mean(cross_score))
```

3) KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)

knn.fit(scale_x_train,y_train)

y_pred_knn = knn.predict(scale_x_test)

print("KNN Training set: ",knn.score(x_train,y_train))
print("KNN Testing set: ",knn.score(x_test,y_test))
print("Accuracy of KNN Classifier: ",accuracy_score(y_test,y_pred_knn))

print("Confusion Matrix:","\n",confusion_matrix(y_test,y_pred_knn))
```

**Cross validation**

```python
from sklearn.model_selection import ShuffleSplit,cross_val_score
knn_model = KNeighborsClassifier()
ssplit = ShuffleSplit(n_splits=10,test_size = 0.20)
cross_score = cross_val_score(knn_model,x,y,cv=ssplit)

print(cross_score)
print(np.mean(cross_score))
```

**4) Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
d_tree = DecisionTreeClassifier()
d_tree.fit(x_train,y_train)

y_pred_tree = d_tree.predict(x_test)

print("Decision tree Training set: ",d_tree.score(x_train,y_train))
print("Decision tree Testing set: ",d_tree.score(x_test,y_test))
print("Accuracy of Decision tree Classifier: ",accuracy_score(y_test,y_pred_tree))

print("Confusion Matrix:","\n",confusion_matrix(y_test,y_pred_tree))
```

**Cross validation**

```python
from sklearn.model_selection import ShuffleSplit,cross_val_score
dec_model = DecisionTreeClassifier()
ssplit = ShuffleSplit(n_splits=10,test_size=0.20,random_state = 0)

cross_score = cross_val_score(dec_model,x,y,cv=ssplit)
print(cross_score)
print(np.mean(cross_score))

```

# **5) XGBoost**

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
xgb = XGBClassifier()
xgb.fit(x_train,y_train)

y_pred_xg = xgb.predict(x_test)

print("xgboost training set:",xgb.score(x_train, y_train))
print("xgboost testing set:",xgb.score(x_test, y_test))
print("Accuracy of xgboost Classifier:", accuracy_score(y_test,y_pred_xg))
print(confusion_matrix(y_test, y_pred_xg))
```

Cross **validation**

```python
from sklearn.model_selection import ShuffleSplit,cross_val_score
xg_model = XGBClassifier()
ssplit = ShuffleSplit(n_splits=10,test_size=0.20,random_state = 0)

cross_score = cross_val_score(xg_model,x,y,cv=ssplit)
print(cross_score)
print(np.mean(cross_score))
```
